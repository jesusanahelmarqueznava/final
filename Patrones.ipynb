```python
import random

# Define productos y sus rangos de ventas mensuales
productos = {
    "Producto A": (100, 200),
    "Producto B": (50, 150),
    "Producto C": (80, 180)
}

# Define período de tiempo en años y meses
años = 3
meses_por_año = 12
total_meses = años * meses_por_año

# Generr datos simulados de ventas mensuales
ventas_mensuales = {}
for producto, rango_ventas in productos.items():
    ventas_mensuales[producto] = [random.randint(rango_ventas[0], rango_ventas[1]) for _ in range(total_meses)]

# Imprime el array de datos simulados
for producto, ventas in ventas_mensuales.items():
    print(f"Ventas mensuales para {producto}: {ventas}")
```

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose

# Crea un DataFrame de pandas con meses como índice
ventas_df = pd.DataFrame(ventas_mensuales, index=pd.date_range(start='2022-01', periods=total_meses, freq='M'))

# Realiza la descomposición de la serie temporal para cada producto
for producto in productos.keys():
    descomposicion = seasonal_decompose(ventas_df[producto], model='additive')

    # Grafica la descomposición
    descomposicion.plot()
    plt.title(f'Descomposición de la serie temporal para {producto}')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Calcula estadísticas descriptivas para cada producto
for producto, ventas in ventas_mensuales.items():
    print(f"Estadísticas de ventas para {producto}:")
    print(f"  Media: {np.mean(ventas)}")
    print(f"  Mediana: {np.median(ventas)}")
    print(f"  Desviación estándar: {np.std(ventas)}")
    print()

# Visualiza la distribución de las ventas a lo largo del tiempo
meses = range(1, total_meses + 1)
for producto, ventas in ventas_mensuales.items():
    plt.plot(meses, ventas, label=producto)

plt.title('Ventas Mensuales')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.legend()
plt.grid(True)
plt.show()
```
